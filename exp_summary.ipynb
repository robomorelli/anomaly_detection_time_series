```python
import pandas as pd
```

```python
df = pd.read_csv('fdir_exp_name.txt', sep=';')
```

```python
display(df.sort_values(['val loss']))
```

```python
df = df.sort_values(['val loss']).drop(columns=['total trial num']).dropna()
```

```python
int_cols = ['num params', 'seq len', 'trial_num']
```

```python
df_str = df.drop(columns=int_cols)
```

```python
df_int = df.dropna().loc[:,int_cols].astype(int)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 100000000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 8)
```

```python
df_tot = df_str.join(df_int)
```

```python
from IPython.display import display
```

```python
'_'.join(['_'.join(df['exp_name'][0].split('_')[:3]), '_'.join(df['exp_name'][0].split('_')[5:])])
```

```python
df_tot['exp_name'] = df['exp_name'].apply(lambda x: '_'.join(['_'.join(x.split('_')[:3])
                                         , '_'.join(x.split('_')[4:])]) if x.startswith('4') else 
                    '_'.join(['_'.join(x.split('_')[:2])
                                         , '_'.join(x.split('_')[3:])]))
```
